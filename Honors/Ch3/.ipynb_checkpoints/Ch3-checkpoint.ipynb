```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import rcParams 
rcParams['text.latex.preamble'] = r'\usepackage{amsmath}' 
rcParams['text.usetex']         = True 
rcParams['font.family']         = 'sans-serif' 
rcParams['font.sans-serif']     = ['Helvetica']

%matplotlib inline
```

### Problem 3.13

```python
def V(x,y,N=10):
    ## unitless: V = V(x,y)/V_0 | x,y = x,y/a 
    V = 0
    for n in range(1,N+1):
        V += np.exp(-(4*n+2)*np.pi*x)*np.sin((4*n+2)*np.pi*y)/(2*n+1)
    V *= 4/np.pi
    return V
```

```python
nx,ny = 100,100
x = np.linspace(0,0.1,nx) # x: 0,infinity
y = np.linspace(0,1,ny) # y: 0,a
X,Y = np.meshgrid(x,y)
```

```python
def plot3d(X,Y,N):
    fig, ax = plt.subplots(1,1,figsize=(7,7),subplot_kw=dict(projection='3d'))

    Z = V(X,Y,N)
    ax.plot_surface(X,Y,Z,cmap=plt.cm.coolwarm)

    ax.set_title(r'$N=%d$'%N,size=20)
    ax.set_xlabel(r'$x/a$',size=20)
    ax.set_ylabel(r'$y/a$',size=20)
    ax.set_zlabel(r'$V/V_0$',size=20)
    
    ax.tick_params(axis='both',which='major',labelsize=20,direction='in')
    ax.view_init(30, 30)
    
    plt.savefig('./plots/prob3-13_%d.png'%N,bbox_inches='tight')
```

```python
N = np.array([1,5,10,20])
for _ in N:
    plot3d(X,Y,_)
```
